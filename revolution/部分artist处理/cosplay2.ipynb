```python
Key_places = {0:0, 7:1, 2:2, 9:3, 4:4, 11:5,
              6:6, 1:7, 8:8, 3:9,10:10, 5:11}
def get_character_vector(dataFrame):
    
    # 获得 artist 的 id
    artists_id = dataFrame["artists_id"].values[0]
#     print(artists_id)
    
    columns = ["danceability","energy","valence","tempo", "loudness",
               "acousticness","instrumentalness","liveness",
               "speechiness","explicit","duration_ms", "mode"]
    popularity_set = dataFrame.loc[:,"popularity"].values + 1
    popularity_total = popularity_set.sum()

    character_vector = []
    for column in columns:
        vector = dataFrame.loc[:,column].values
        mean = ((popularity_set * vector) / popularity_total).sum()
        character_vector.append(mean)
    
    key_x, key_y = [], []
    key_set = dataFrame.loc[:,"key"].values
    for i in range(len(popularity_set)):
        theta = math.pi / 6 * Key_places[key_set[i]]
        key_x.append(math.sin(theta))
        key_y.append(math.cos(theta))
    key_x = np.array(key_x)
    key_y = np.array(key_y)
    character_vector.append(((key_x * popularity_set) / popularity_total).sum())
    character_vector.append(((key_y * popularity_set) / popularity_total).sum())

    return np.array(character_vector), artists_id
    
get_character_vector(subFrame)
```

```python
def get_character_difference(dataFrame, x_0 = leg):
    # 获得 artist 的 id
    artists_id = dataFrame["artists_id"].values[0]
#     print(artists_id)
    
    columns = ["danceability","energy","valence","tempo", "loudness",
               "acousticness","instrumentalness","liveness",
               "speechiness","explicit","duration_ms", "mode"]
    popularity_set = dataFrame.loc[:,"popularity"].values + 1
    works_number = len(popularity_set)
    x_axis = np.arange(works_number)
    power = (x_axis - (works_number - 1)/2 )/ x_0
    weight_start = 1 / (np.exp(power) + 1)
    weight_end = 1 / (np.exp(-power)+1)
    
#     print(weight_start)
#     print(weight_end)
    
    
    vector_start_list = []
    vector_end_list   = []

    for column in columns:
        vector = dataFrame.loc[:,column].values
        start_mean = (popularity_set * vector * weight_start).sum() / (popularity_set * weight_start).sum()
        end_mean   = (popularity_set * vector * weight_end  ).sum() / (popularity_set * weight_start).sum()
        vector_start_list.append(start_mean)
        vector_end_list.append(end_mean)
    
    key_x, key_y = [], []
    key_set = dataFrame.loc[:,"key"].values
    for i in range(len(popularity_set)):
        theta = math.pi / 6 * Key_places[key_set[i]]
        key_x.append(math.sin(theta))
        key_y.append(math.cos(theta))
    key_x = np.array(key_x)
    key_y = np.array(key_y)
    key_x_mean = (key_x * popularity_set * weight_start).sum() / (popularity_set * weight_start).sum()
    key_y_mean = (key_y * popularity_set * weight_start).sum() / (popularity_set * weight_start).sum()

    vector_start_list.append(key_x_mean)
    vector_start_list.append(key_y_mean)
    vector_end_list.append(key_x_mean)
    vector_end_list.append(key_y_mean)

    
    vector_start = np.array(vector_start_list)
    vector_end   = np.array(vector_end_list)

    return vector_start,vector_end, artists_id


get_character_difference(subFrame)
```

```python
def cos_13D(vect1, vect2):
    len1 = np.sqrt(np.sum(vect1 ** 2))
    len2 = np.sqrt(np.sum(vect2 ** 2))
    return np.sum(vect1 * vect2) / (len1 * len2)
```

```python
artist_number = 5641

artist_id_list = []
artist_trait_dict = {}
artist_trait_list = []
artist_trait_start_dict = {}
artist_trait_start_list = []
artist_trait_end_dict   = {}
artist_trait_end_list   = []

for num, artist_range in tqdm(enumerate(range_list),):
    subFrame = extractedData.loc[artist_range[0]:artist_range[1]-1]
    subFrame = subFrame.sort_values(by=["artists_id", 'year'], ascending=True)
    character_vector, artist_id = get_character_vector(subFrame)
    vector_start,vector_end, artist_id = get_character_difference(subFrame)
    
    artist_id_list.append(artist_id)
    
    artist_trait_list.append(character_vector)
    artist_trait_start_list.append(vector_start)
    artist_trait_end_list.append(artist_id)
    
    artist_trait_dict[artist_id] = character_vector
    artist_trait_start_dict[artist_id] = vector_start
    artist_trait_end_dict[artist_id] = vector_end
    

artist_trait = np.array(artist_trait_list)
artist_trait_start = np.array(artist_trait_start_list)
artist_trait_end   = np.array(artist_trait_end_list)

```

```python
artist_number = 1154
influence = np.zeros([artist_number,artist_number])
for i in tqdm(range(artist_number)):
    influencer = artist_id_list[i]
    influencer_character = artist_trait_dict[influencer]
    for j in range(artist_number):
        influenced = artist_id_list[j]
        influencer_start = artist_trait_start_dict[influenced]
        influencer_end = artist_trait_end_dict[influenced]
        
        cos_theta = cos_13D(influencer_character-influencer_start, influencer_end-influencer_start)
        influence[i][j] = cos_theta
```

```python
df = pd.DataFrame(influence)
df.to_csv("mini_influence_relationship.csv")
```
