# 首先从influence data当中获得一个 artists 到genra 的映射

```python
influence_col = influence_data.loc[:,["influencer_id","influencer_name","influencer_main_genre","influencer_active_start"]]
influence_col.columns =["id","name","genre","year"]
influence_col.head()
follower_col = influence_data.loc[:,["follower_id","follower_name","follower_main_genre","follower_active_start"]]
follower_col.columns =["id","name","genre","year"]
person_col = pd.concat([influence_col,follower_col])

# per_col.index=range(85540)
person_col.shape
person_col.head()
```

```python
per_map=person_col.drop_duplicates(["id"])
per_map.head()
```

```python
def getGenre(id):
    try:
        return per_map[per_map.id==id].iloc[0,2]
    except:
#         print(id)
        return "Unknown"
getGenre(12138)
```

```python
# build a dictionary
Genre_to_id = {}
cnt = 0

# print(per_map["name"])
for idx, row in per_map.iterrows():
    name = per_map["name"][idx]
    if type(name) == str:
        Genre_to_id[idx] = name
        print("Normal")
    else:
        print(f"[{cnt}] type {type(name)}")
        print(row)
        cnt+=1

    
f = open("Genre_to_id_dictionary.txt", "w")
f.write("{\n")
for i, key in enumerate(Genre_to_id):
    f.write('{}:"{}",\n'.format(key,Genre_to_id[key]))
f.write("}\n")
f.close()
```

## 建立一个字典genre_map,将数字对应到不同的流派

```python
genre_map=influence_data.drop_duplicates(["follower_main_genre"])["follower_main_genre"]
genre_map.index=range(20)
genre_map=genre_map.to_dict()
genre_map # genre对应的字典
```

# 建立一个字典re_genre_map，将流派对应回数字

```python
re_genre_map={}
for i,key in enumerate(genre_map):
    re_genre_map[genre_map[key]]=key
re_genre_map
```

```python
dataFrame_full_artist.head()
```

```python
tmp=0
for i,r in tqdm(dataFrame_full_artist.iterrows(),):
    m=r["artists_id"].find(',')
    Id=0
    if m!=-1:
        Id=int(r["artists_id"][1:m])
    else:
        Id=int(r["artists_id"][1:-1])
    dataFrame_full_artist.loc[i,"genre"]=getGenre(Id)
    if dataFrame_full_artist.loc[i,"genre"]!=-1:
        dataFrame_full_artist.loc[i,"genre_id"]=re_genre_map[dataFrame_full_artist.loc[i,"genre"]]
    else:
        dataFrame_full_artist.loc[i,"genre_id"]=-1

dataFrame_full_artist
```

```python
dataFrame_full_artist = dataFrame_full_artist.sort_values(by=['genre_id', 'year', 'artists_id'], ascending=True)
dataFrame_full_artist
```

```python
genre_number = dataFrame_full_artist["genre_id"].value_counts()
value = genre_number.values
index = genre_number.index.values
genre_dict = {}
for i in range(len(value)):
    genre_dict[index[i]] = value[i]
    
genre_num = []
for i in range(len(value)):
    genre_num.append([i,genre_dict[i]])
genre_num = np.array(genre_num).T[-1]
genre_position = np.cumsum(genre_num)
print("genre_num = \n", genre_num)
print("genre_position = \n", genre_position)
```
